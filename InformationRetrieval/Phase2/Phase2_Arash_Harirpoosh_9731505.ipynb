```python
import json
import pandas as pd
from Phase2.vectorization import Vectorization
```

Read the documents

```python
df = pd.read_excel('../data/IR1_7k_news.xlsx')
df_content = df['content']
df_title = df['title']
vect = Vectorization(df_title, df_content, df.shape[0])
```

Calculate and save the tfidf score for all of the tokens in documents

```python
with open('../Phase1/positionalIndex.json') as json_file:
    positional_index = json.load(json_file)
vect.calculate_all_docs_tfidf(positional_index=positional_index)
vect.save_positional_index()
```

Read the calculated indexes and scores

```python
vect.read_positional_index()
```

Calculate and save the champion list

```python
vect.construct_champion_list(10)
vect.save_champion_list()
```

Read the constructed champion list

```python
vect.read_champion_list()
```

# Train Word2Vec Model

## Train model

```python
td = vect.prepare_training_data()
vect.save_pickle_data(td, 'training_data.obj')
vect.train_word2vec_model(vect.load_pickel_data('training_data.obj'))
```

## Load Model

```python
model = vect.load_w2v_model()
all_docs_vec = vect.generate_docs_vector()
vect.save_pickle_data(all_docs_vec, 'docs_vec.obj')
```
