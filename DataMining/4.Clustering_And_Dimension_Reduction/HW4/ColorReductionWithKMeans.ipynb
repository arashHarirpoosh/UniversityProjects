```python
from HW4.KMeans import *
from matplotlib import image
import matplotlib.pyplot as plt
```

Read The Image And Convert It To The 2D Array

```python
img = image.imread('data/sample_img1.png')
dim_one = img.shape[0] * img.shape[1]
dim_two = img.shape[2]
img_2d = img.reshape(dim_one,dim_two)
```

Use KMeans To Cluster The 2D Image Colors And Then Replace The RGB Values With The Centroid RGB

```python
K = [2, 4, 16, 32, 64]
for k in K:
    centroids, clusters = k_mean(img_2d, k)
    # Update The RGB Values Of The Image
    compressed_img = img.copy()
    dim0, dim1 = img.shape[:2]
    for i in range(dim0):
        for j in range(dim1):
            min_dist = float('inf')
            selected_cluster = None
            for c in range(k):
                dist = np.linalg.norm(img[i][j] - centroids[c])
                if dist < min_dist:
                    min_dist = dist
                    selected_cluster = c
            compressed_img[i][j] = centroids[selected_cluster]
    image.imsave('compressed_img_'+str(k)+'.png', compressed_img)
    plt.imshow(compressed_img)
    # plt.savefig('compressed_img_'+str(k)+'.png')
    plt.show()
```
